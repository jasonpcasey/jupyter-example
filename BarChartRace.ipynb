```python
# Bar chart Race
park_year_data = df.groupby(["ParkName", "Year"]).agg({"RecreationVisits": "sum"}).reset_index()

park_year_data = park_year_data.sort_values(by=["Year", "RecreationVisits"], ascending=[True, False])

fig = px.bar(
    park_year_data,
    x="RecreationVisits",
    y="ParkName",
    color="ParkName",
    orientation="h",
    animation_frame="Year",
    title="National Park Visits Over Time",
    labels={"RecreationVisits": "Visits", "ParkName": "Park"},
    height=800,
)

# fig.update_layout(yaxis=dict(autorange="reversed"))
fig.update_layout(
    xaxis_title="Number of Visits",
    yaxis_title="Parks",
    xaxis={'categoryorder':'total descending'},
    showlegend=False,
    margin={"r": 0, "t": 50, "l": 150, "b": 0},
)

fig.show()
```
